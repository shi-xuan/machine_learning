```python
#畫出 sin wave
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
pt.xlim(0,360)     #x軸範圍
pt.ylim(-1.2,1.2)    #y軸範圍
pt.title("Sin function")  #標題
pt.plot(x,y)
pt.show()
```

```python
#畫出 sin & cos wave
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)   #180頻率(弧度)
z=np.cos(x*np.pi/180.0)   #180頻率(弧度)
pt.xlim(0,360)     #x軸範圍
pt.ylim(-1.2,1.2)    #y軸範圍
pt.title("Sin&Cos function")  #標題
pt.plot(x,y, color="lightblue")  #sin顏色
pt.plot(x,z, color="pink")   #cos顏色
pt.xlabel("x")  #x軸
pt.ylabel("y")  #y軸
pt.show()
```

```python
#3D plots
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
from matplotlib import cm

def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,
                y,
                f(x, y),
                rstride=2, cstride=2,
                cmap=cm.jet,
                alpha=0.7,
                linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```
